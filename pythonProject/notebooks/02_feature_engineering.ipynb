```python
import pandas as pd
df = pd.read_csv(r'../data/results.csv')

df['date'] = pd.to_datetime(df['date'])
df = df[df['tournament'] != 'Friendly']

def winner(row):
    if(row['home_score'] > row['away_score']):
        return 'home'
    elif(row['home_score'] < row['away_score']):
        return 'away'
    else:
        return ('draw')

df['winner'] = df.apply(winner, axis=1)
print(df.head(5))

home_df = pd.DataFrame({
    'date': df['date'],
    'team': df['home_team'],
    'won': df['winner'] == 'home'
})

away_df = pd.DataFrame({
    'date': df['date'],
    'team': df['away_team'],
    'won': df['winner'] == 'away'
})

win_result = pd.concat([home_df, away_df])
win_result = win_result.reset_index(drop=True)
print(win_result.head(5))
print(win_result.tail(5))
print(len(win_result))
print(len(df))

win_rate = win_result.groupby('team')['won'].mean()
print(win_rate)
date_sorted = win_result.sort_values('date')
last_10 = date_sorted.groupby('team').tail(n=10)
win_rate_10 = last_10.groupby('team')['won'].mean()
print(win_rate_10)
print(win_rate_10['Brazil'])

team_features = pd.concat([win_rate, win_rate_10], axis=1)
team_features.columns = ['overall_win_rate', 'recent_win_rate']
print(team_features)
```
